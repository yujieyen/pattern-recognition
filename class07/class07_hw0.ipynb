```python
# 淺嚐 pandas 
# 運用 yfinance 撈股票價格歷史資料 
 

import pandas as pd 
import yfinance as yf 
 

# 選3支股票，世界前 3大。 
tks= yf.Tickers(['MSFT', 'AAPL', '2222.SR','GOOG','AMZN','TSLA','FB','NVDA','BRK-A','TSM']) 
tks 
# 撈1年的資料 
tkPrice= tks.history(period='5y') 
 

# 簡單作圖 X= X(t) 
tkPrice.Close.plot() 
tkPrice.Close.plot(logy= True) 
 

# Histogram 
tkPrice.Close.hist(orientation= 'horizontal') 

 

#statistical description 
tkPrice.Close.describe() 
```
